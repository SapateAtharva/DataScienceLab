```
# Practical 2: Text Summarization using NLP

# Import Libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from heapq import nlargest

# Download required data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Sample text (e.g., Government Policy Document)
text = """India’s vaccination drive against COVID-19 is one of the largest in the world.
The government has implemented multiple phases to ensure equitable distribution.
Health workers and elderly citizens were prioritized.
Technology platforms like CoWIN facilitated online registration and tracking.
As a result, vaccination rates have steadily improved, contributing to the country’s pandemic recovery."""

# Tokenization
sentences = sent_tokenize(text)
words = word_tokenize(text.lower())

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalnum() and word not in stop_words]

# Calculate word frequencies
freq = nltk.FreqDist(filtered_words)
max_freq = max(freq.values())
for word in freq.keys():
    freq[word] = (freq[word]/max_freq)

# Score each sentence
sentence_scores = {}
for sent in sentences:
    for word in nltk.word_tokenize(sent.lower()):
        if word in freq:
            sentence_scores[sent] = sentence_scores.get(sent, 0) + freq[word]

# Select top sentences for summary
summary_sentences = nlargest(2, sentence_scores, key=sentence_scores.get)
summary = ' '.join(summary_sentences)

print("\n--- Original Text ---\n", text)
print("\n--- Summary ---\n", summary)
```
